# **IMAGE TO PENCIL SKETCH WITH PYTHON**

## Import Libraries

```python
import cv2
import matplotlib.pyplot as plt
```

## Read the Image

```python
img=cv2.imread("/content/man-of-steel.jpg")
```

## Display the Original Image

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

## Transforming the Image into a Pencil Sketch

### Convert OG Image to Gray-Scale Image

```python
grey_img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(grey_img)
```

### Invert the Gray-Scale Image

```python
invert_img=cv2.bitwise_not(grey_img)
cv2_imshow(invert_img)
```

### Blur the Inverted Image

```python
blur_img=cv2.GaussianBlur(invert_img, (111,111),0)
cv2_imshow(blur_img)
```

### Invert the Blurred Image

```python
invblur_img=cv2.bitwise_not(blur_img)
cv2_imshow(invblur_img)
```

### Pencil Sketch Image

```python
sketch_img=cv2.divide(grey_img,invblur_img, scale=256.0)
cv2.imwrite('sketch.png', sketch_img)
cv2_imshow(sketch_img)
```
