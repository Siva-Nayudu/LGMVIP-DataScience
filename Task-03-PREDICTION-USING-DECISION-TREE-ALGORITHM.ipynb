## Import Necessary Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Load Dataset

```python
df = pd.read_csv("Iris.csv")
df.head()
```

## Data Summarization

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
x=df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y=df[['Species']]
```

## Visualizing Data

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')
colors = ['red','blue','green']
sp = df['Species'].unique()
ct = df['Species'].value_counts().tolist()
ax.pie(ct, labels = sp, autopct='%1.2f%%', colors=colors, shadow=True, startangle=90)
plt.show()
```

## Fitting Data into Model

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier()
clf.fit(x,y)
```

## Import Libraries for Data Visualization

```python
from six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
```

## Visualize Model as Graph

```python
features = df.columns[:-1]
classes = df['Species'].unique().tolist()
```

```python
from sklearn.tree import plot_tree
plt.figure(figsize=(20, 15))
plot_tree(clf, feature_names=features, class_names=classes, filled=True)
plt.show()
```
